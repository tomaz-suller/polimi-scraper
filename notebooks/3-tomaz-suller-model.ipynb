## Model

```python
m = mip.Model()
```

### Domain

```python
time_domain = occupancy_df.index.to_list()
room_domain = occupancy_df.columns.to_list()
print(time_domain[:5], time_domain[-5:])
print(room_domain[:5], room_domain[-5:])
```

### Inputs

```python
bx: pd.Series = classroom_df["x_building"].loc[SEDE]
by: pd.Series = classroom_df["y_room"].loc[SEDE]
f: pd.Series = classroom_df["floor"].loc[SEDE]
rx: pd.Series = classroom_df["x_room"].loc[SEDE]
ry: pd.Series = classroom_df["y_room"].loc[SEDE]
```

```python
o = occupancy_df.astype("int")
o.head()
```

```python
s = schedule_df.astype("int")
s.head()
```

```python
M = 1_000
```

### Decision variables

```python
p: dict[Any, dict[Any, mip.Var]] = {}
delta_p = {}
d_bx = {}
d_by = {}
d_f = {}
d_rx = {}
d_ry = {}
```

```python
for t in time_domain:
    p_t = {}
    for r in room_domain:
        p_t[r] = m.add_var(f"p_{r}_{t}", var_type=mip.BINARY)
    p[t] = p_t

    if t == time_domain[0]:
        continue
    d_bx[t] = m.add_var(f"dBx_{t}", var_type=mip.CONTINUOUS)
    d_by[t] = m.add_var(f"dBy_{t}", var_type=mip.CONTINUOUS)
    d_f[t] = m.add_var(f"dF_{t}", var_type=mip.CONTINUOUS)
    d_rx[t] = m.add_var(f"dRx_{t}", var_type=mip.CONTINUOUS)
    d_ry[t] = m.add_var(f"dRy_{t}", var_type=mip.CONTINUOUS)
```

### Constraints

A student can only be in a single room at any time

```python
for t in time_domain:
    m += mip.xsum(p[t][r] for r in room_domain) == 1, f"uniqueness_{t}"
```

A student must attend their lectures

```python
for t in time_domain:
    for r in room_domain:
        m += p[t][r] >= s.loc[t, r], f"attendance_{t}_{r}"
```

A student must be in a room that is free, except when the student has a lecture in that room

```python
for t in time_domain:
    for r in room_domain:
        m += p[t][r] <= o.loc[t, r] + s.loc[t, r], f"occupancy_{t}_{r}"
```

#### Auxiliary distance variables

We use auxiliary variables to define the objective function instead of optimising the distances directly so that we can "toggle" distances

We use two modelling tricks to frame the situation as a mixed integer linear programming problem: `TODO`

##### Building distance $d_{B_x}$ and $d_{B_y}$

```python
for t in time_domain[1:]:
    for b, d_b in zip((bx, by), (d_bx, d_by)):
        building_distance = mip.xsum(
            b.loc[r] * (p[t][r] - p[t - 1][r]) for r in room_domain
        )
        m += d_b[t] >= building_distance
        m += d_b[t] >= (-1) * building_distance
```

##### Floor distance $d_F$

```python
for t in time_domain[1:]:
    floor_distance = mip.xsum(f.loc[r] * (p[t][r] - p[t - 1][r]) for r in room_domain)
    same_building_factor = mip.xsum(
        M * bx.loc[r] * by.loc[r] * (p[t][r] - p[t - 1][r])
        for r in room_domain
    )
    m += d_f[t] >= floor_distance - same_building_factor
    m += d_f[t] >= (-1) * floor_distance - same_building_factor
```

##### Room distance $d_{R_x}$ and $d_{R_y}$

```python
for t in time_domain[1:]:
    for r_, d_r in zip((rx, ry), (d_rx, d_ry)):
        room_distance = mip.xsum(
            r_.loc[r] * (p[t][r] - p[t - 1][r]) for r in room_domain
        )
        same_floor_and_building_factor = mip.xsum(
            M * f.loc[r] * bx.loc[r] * by.loc[r] * (p[t][r] - p[t - 1][r])
            for r in room_domain
        )
        m += d_r[t] >= room_distance - same_floor_and_building_factor
        m += d_r[t] >= (-1) * room_distance - same_floor_and_building_factor
```

There are no type hints on the overloaded implementation of `+=` for `mip.Model`, so the language server fails to recognise `m` as `mip.Model`, and we add the type hint to address that

```python
m: mip.Model
```

### Objective function

```python
m.objective = mip.minimize(
    mip.xsum(
        d_bx[t] + d_by[t] + d_f[t] + d_rx[t] + d_ry[t]
        for t in time_domain[1:]
    )
)
```

## Optimise

```python
status = m.optimize()
status
```
