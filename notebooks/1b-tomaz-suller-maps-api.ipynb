```python
from collections import defaultdict

import pandas as pd
import requests
from tqdm.notebook import tqdm

from polimi_scraper.config import logger, DataPath
```

```python
URL = "https://maps.polimi.it/maps/infocadmanagement/controller/google/MapManager.do"
CONSTANT_PARAMETERS = {
    "evn_CARICA_ENTITA2": "evento",
    "bounds": "-90,-180,90,180",
}
ZOOM_LEVEL_RANGE = range(11, 16+1)
POINT_OF_INTEREST_KEY_PREFIX = "g_"
```

```python
response_dict = defaultdict(list)
for zoom_level in tqdm(ZOOM_LEVEL_RANGE):
    params = {"zoom": zoom_level} | CONSTANT_PARAMETERS
    response = requests.get(URL, params=params, timeout=5)
    if response.status_code == requests.codes.ok:
        for key, value in response.json().items():
            if key.startswith(POINT_OF_INTEREST_KEY_PREFIX):
                response_dict[key.replace(POINT_OF_INTEREST_KEY_PREFIX, "")].extend(value)
    else:
        logger.error(f"Error for zoom level {zoom_level}: {response.status_code}")
response_dict
```

```python
response_df_dict = {
    key: pd.json_normalize(value).drop_duplicates(subset="codice_patrimonio")
    for key, value in response_dict.items()
}
for key, value in response_df_dict.items():
    if value.empty:
        continue
    print(key)
    display(value)
```

```python
for property_type, df in response_df_dict.items():
    if df.empty:
        continue
    df.to_parquet(DataPath.RAW_PLACES_DIR / f"{property_type}.parquet", index=False)
```
